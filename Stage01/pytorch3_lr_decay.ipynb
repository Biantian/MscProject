```
import matplotlib.pyplot as plt
```

```
def exponential_LR(lr,epoch,gamma=0.95):
  # Args:
  #   param: lr: Initial learning rate.
  #   param: gamma: Multiplicative factor of learning rate decay.
  #   param: epoch: Total number of epoch.
  result = [lr*pow(gamma,i) for i in range(epoch)]
  return result

def lambda_LR(lr,epoch,gamma=0.95,skip=5):
  # Args:
  #   param: lr: Initial learning rate.
  #   param: gamma: Multiplicative factor of learning rate decay.
  #   param: epoch: Total number of epoch.
  #   param: skip: Skip the first n epoch.
  result = [lr]*skip
  for i in range(skip,epoch):
    result.append(lr*pow(gamma,i-skip))
  return result
```

```
x = list(range(30))
y = exponential_LR(0.001,30,0.9)
y2 = lambda_LR(0.001,30,0.9,5)

print(x)
print(y)
print(y2)
```

```
_,ax = plt.subplots()
ax.plot(x,y,'b-',label='ExponentialLR')
ax.plot(x,y2,'y--',label='LambdaLR')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.legend()
```
